```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\anokh\OneDrive\Desktop\veracity\sales_data (2).csv")
```

```python
df.head(20)
```

```python
# fill missing discounts with 0
df['Discount'] = pd.to_numeric(df['Discount'], errors = 'coerce')
df['Discount'] = df['Discount'].fillna(0)
```

```python
df.head(20)
```

```python
## add new column
df['Revenue'] = df['Price'] * df['Quantity']
```

```python
df.head(10)
```

```python
# apply discount ---final revenue
df['Final_Revenue'] = df['Revenue'] * (1-df['Discount'])
```

```python
df.head(10)
```

```python
# filter
electronics_df = df[df['Category'] == 'Electronics']
print("Electronics only:\n", electronics_df.head(),"\n")
```

```python
# group by category --- total sales
category_sales = df.groupby("Category")["Final_Revenue"].sum()
print("Total sales by category:\n", category_sales,"\n")
```

```python
df.head(10)
```

```python
# sorting
top_products = df.groupby("Product")["Final_Revenue"].sum().sort_values(ascending= False).head(5)
print("Top 5 Products By Revenue:\n", top_products, "\n")
```

```python
summary_stats = df.describe()
print("Summary Statistics:\n", summary_stats, "\n")
```

```python
df.head(200)
```

```python
# Filter
high_sales = df[df["Final_Revenue"] > 5000]
print("Products with Final Revenue > 5000:\n", high_sales,"\n")
```

```python
df.head(15)
```

```python
sorted_df = df.sort_values(by=["Category","Final_Revenue"],ascending =[True,False])
print("Sorted by category & Final Revenue:\n", sorted_df.head(),"\n")                           
```

```python
# find top 3 salees per category
top3_per_category = (df.sort_values("Final_Revenue", ascending= False).groupby("Category").head(3))
print("Top 3 products per category:\n",top3_per_category)
```

```python
# pivot table--- (cate vs avg revenue & quantity)
pivot_table = pd.pivot_table(df, values =["Final_Revenue", "Quantity"],
                             index="Category",
                             aggfunc = {"Final_Revenue": "mean","Quantity":"sum"})
print("Pivot Table:\n",pivot_table,"\n")
```

```python
# crosstab (discount usage frequency per category)
discount_crosstab = pd.crosstab(df["Category"],df["Discount"])
print("Discount Frequency by Category:\n",discount_crosstab,"\n")
```

```python
df.head(10)
```

```python
# create a new column -- high/low sales
df["Sales_Flag"] = df["Final_Revenue"].apply(lambda x: "High" if x > 3000 else "Low")
print("Data with Sales_Flag column:\n", df.head(),"\n")
```
