```python
# covariance
- it measures the direction of the linear relationship between two variables
+covariance --- when one variable increases, the other tends to increase
-Covariance --- when one variable increases, the other tends to decrease
zero covariance -- no linear relationship 

#Corrrelation
 it the stadardized version of covariance that tells us how strong in what direction two variables are related
Range - Always between -1 to +1
1. +1 -- perfect positive relationship(as one goes up , other always goes up in fixed proportion)
2. -1 --perfect negative relationship(as one goes up, the other always goes down in fixed proportion)
3. 0-- No linear relationship


# Probability
It simply measure the how likely an event is to happen:
0 --- event is impossible
1 --- event is certain

you you can also express it as a percentage(0% to 100%)
```

```python
Scenario:
ice cream sales vs sunglasses sales & weather

Step 1
you own a beach shop that sells : ice cream & sunglassses
you also have weather data(sunny, cloudy, rainy)

Step 2
probability
you look at previous daat and find
P(sunny)= 0.6
p(Cloudy) = 0.3
P(rainy)= 0.1

Interpretation - probability tells you how likely certain sales levels are, given the weather

Step 3
Covariance

over the season you record:
On sunny days --- both icecream and sunglasses sales rise
on rainy days ---- both drop

whhhen you compute covariance, it is positive--- meaning they generally move in the same direction (up together , down together)

Step4
Correlation

you standardized that covariance and find:
correlation = 0.95% --- very strong positive relationship

This means-- when ice cream sales are high, sunglasses sales are almost too high, and vice versa.

insights:
Probability - helps you predict the likelihood of each sales scenario based on weather
Covariance -- both pproduct tend to move in the same direction
correlation -- tells you how strong that movement is

business takeway:
if the weather forecast predicts sunny weekends
1. high probability of high sales for bothy products
covariance confirms they rise together
3. high correlation suggest bundling sunglass + icecream could be smart promo.
```

```python
# Bayes Theorem
how to update the probability of something being true when you get new evidence
used - statistics, Machine Learning, and Spam detection

Prior - before hearing any clues, how likely is it the suspect is guilty?
Likelihood; If the suspect were guilty, how likely is it they would leave this evidence?
Evidence : how often does this evidence show up in general?
Posterior: After consider the clue, how likely is it now that they are guilty?
```
